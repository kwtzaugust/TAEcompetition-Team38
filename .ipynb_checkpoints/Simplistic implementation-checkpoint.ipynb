# Cleaning

```r
library(RWeka)
library(tm)
```

```r
corpus <- VCorpus(VectorSource(data$tweet))
corpus <- tm_map(corpus,content_transformer(tolower))
corpus <- tm_map(corpus,removeWords,stopwords("english"))
# corpus <- tm_map(corpus,removeWords,c("weather", "mention"))
corpus <- tm_map(corpus,removePunctuation)
corpus <- tm_map(corpus,removeNumbers)
corpus <- tm_map(corpus,stemDocument)
# corpus <- tm_map(corpus,removeWords,c("raini", "now", "day"))
```

```r
BigramTokenizer <- function(x) NGramTokenizer(x, Weka_control(min = 1, max = 2))
dtm <- DocumentTermMatrix(corpus, control = list(tokenize = BigramTokenizer))
```

```r
inspect(dtm[1,])
```

```r
findFreqTerms(dtm,lowfreq=50)
```

```r
dtm_sparsity = 0.995
dtm <- removeSparseTerms(dtm,dtm_sparsity)
```

```r
findFreqTerms(dtm,lowfreq=50)
```

```r
twittersparse <- as.data.frame(as.matrix(dtm))
head(twittersparse)
```

```r

library(wordcloud)
word_freqs = sort(colSums(twittersparse), decreasing=TRUE) 
# Create data frame with words and their frequencies
dm = data.frame(word=names(word_freqs), freq=unname(word_freqs))
# Plot wordcloud
wordcloud(dm$word, dm$freq, random.order=FALSE, colors=brewer.pal(8, "Dark2"))
```

# Split

```r
library(caTools)
set.seed(1)
spl <- sample.split(twittersparse$sentiment, SplitRatio=0.7)
train <- subset(twittersparse,spl==TRUE)
test <- subset(twittersparse,spl==FALSE)
table(train$sentiment)
```

```r
head(train)
```

# Clusters

```r
limitedTrain = train
limitedTrain$sentiment = NULL
limitedTest = test
limitedTest$sentiment = NULL
```

```r
library(caret)
preproc = preProcess(limitedTrain)
normTrain = predict(preproc, limitedTrain)
normTest = predict(preproc, limitedTest)
```

```r
mean(normTrain$lol)
mean(normTest$lol)
```

```r
set.seed(100)
centers = 2
km = kmeans(normTrain, centers = centers , nstart = 20)
table(km$cluster)
```

```r
library(flexclust)
km.kcca = as.kcca(km, normTrain)
clusterTrain = predict(km.kcca)
clusterTest = predict(km.kcca, newdata = normTest)
table(clusterTrain)
table(clusterTest)
```

```r
for (k in 1:centers) {
    name1 = paste0("train",k)
    value1 = subset(train, clusterTrain == k)
    assign(name1, value1)
    name2 = paste0("test",k)
    value2 = subset(test, clusterTest == k)
    assign(name2, value2)
}
```

# CART

```r
library(rpart)
set.seed(1)

for (k in 1:centers) {
    train_data = eval(parse(text = paste0("train",k)))
    assign(paste0("CART",k), rpart(as.factor(sentiment)~., data=train_data, cp=1e-10))
}
```

```r
printcp(CART13)
```

```r
CART6 <- prune(CART6,cp=1e-02)
CART9 <- prune(CART9,cp=9.0481e-05)
CART11 <- prune(CART11,cp=1.5e-02)
CART12 <- prune(CART12,cp=1e-03)
```

Minimum xerror happens around 9.8261e-05

### Predictions

```r
for (k in 1:centers) {
    model = eval(parse(text = paste0("CART",k)))
    newdata = eval(parse(text = paste0("test",k)))
    assign(paste0("CARTpredict",k), predict(model,newdata=newdata,type="class"))
}


AllPredictions = NULL
AllOutcomes = NULL
for (k in 1:centers) {
    prediction = eval(parse(text = paste0("CARTpredict",k)))
    AllPredictions = c(AllPredictions, prediction)
    outcome = eval(parse(text = paste0("test",k, "$sentiment")))
    AllOutcomes = c(AllOutcomes, outcome)
}

CM<-table(AllPredictions, AllOutcomes)  # chack with lower values
CM
Accuracy = (CM[1,1]+CM[2,2]+CM[3,3])/sum(CM)
Accuracy
```

|dtm_sparsity |accuracy|
|-|-|
|0.990|79%|
|0.993|80%|
|0.995|80%|

### Evaluation

```r
for (k in 1:centers) {
    print(paste0("Accuracy of cluster ",k))
    prediction = eval(parse(text = paste0("CARTpredict",k)))
    outcome = eval(parse(text = paste0("test",k, "$sentiment")))
    CM<-table(prediction, outcome)  # chack with lower values
    print(CM)
}
```

# Random Forest

```r
library(randomForest)
set.seed(1)
ntree = 200
nodesize = 10

for (k in 1:centers) {
    train_data = eval(parse(text = paste0("train",k)))
    assign(paste0("forest",k), randomForest(as.factor(sentiment)~., data=train_data, nodesize=nodesize, ntree=ntree))
}
```

### Predictions

```r
# forestpredict <- predict(forest,newdata=test,type="class")
# CM<-table(forestpredict, test$sentiment)  # chack with lower values
# CM
# Accuracy = (CM[1,1]+CM[2,2]+CM[3,3])/sum(CM)
# Accuracy


for (k in 1:centers) {
    model = eval(parse(text = paste0("forest",k)))
    newdata = eval(parse(text = paste0("test",k)))
    assign(paste0("forestpredict",k), predict(model,newdata=newdata,type="class"))
}


AllPredictions = NULL
AllOutcomes = NULL
for (k in 1:centers) {
    prediction = eval(parse(text = paste0("forestpredict",k)))
    AllPredictions = c(AllPredictions, prediction)
    outcome = eval(parse(text = paste0("test",k, "$sentiment")))
    AllOutcomes = c(AllOutcomes, outcome)
}

CM<-table(AllPredictions, AllOutcomes)  # chack with lower values
CM
Accuracy = (CM[1,1]+CM[2,2]+CM[3,3])/sum(CM)
Accuracy
```
